### Introduction

In this project, we use Stochastic Differential Equations, Ito's lemma, and numerical schemes for SDEs to solve the Black Scholes Option Pricing SDE. We introduce concepts of Euler Maruyama and Milstein scheme which are specific to numerically solving SDEs. The derivations of these concepts are provided and the discretization of the Black Scholes model is provided for both numerical schemes.  SDEs are equations used to define stochastic, or random processes in nature. By definition, these equations contain atleast one random component, usually defined by a Brownian motion. A standard SDE is defined below:
$$dX = \mu(t,X)dt + \sigma(t,X)dB \ \ \ \ \ (1)$$

where X is a stochastic process, ùúá is the drift term of the process, ùúé is the diffusion term, ùë° is the time and ùêµ is a Brownian motion. The Black Scholes model describes the price of an option as a geometric Brownian motion, which is used to derive the SDE. The drift, $\mu$ and diffusion, $\sigma$ are replaced with the risk- free interest rate and volatility respectively. The Black-Scholes SDE is defined as:

$$dX = r(t, X)dt + \sigma(t, X)dB \ \ \ \ \ (2)$$

Ito's lemma is used to derive the Ito-Taylor expansion for an SDE. The Euler-Maruyama and Milstein‚Äôs numerical approximation methods of an SDE by truncating the Ito-Taylor expansion are derived. Finally, these two techniques are applied to a Black-Scholes  SDE to simulate the numerical solutions while comparing their solutions to the analytic solution of a Black-Scholes SDE. A comparative study is carried out in the last section

Some terms to keep note of:

1. Call Option: This is an option to buy a security on or before a specified time known as exercise date.
2. European Call Option: This is an option to buy a security on a specified exercise date.

Brownian Motion $B_t$,  $0 \leq t \leq T$ :-
This is a set of random variables, one for each value of the real variable t in the interval [0, T] which possess the following properties:

1. $B_t$ is continuous in the parameter $t$

2. For each $0 \leq s \leq t, B_t ‚àí B_s$ is normally distributed with $B_t = 0$, $Var(B_t) = t$ and $B_t, B_s$ are independent of each other.

3.	For each $t$ and $s$, the random variables $B_{t+s} ‚àí B_s$ and $B_s$ are independent. Moreover $B_{t+s} ‚àí B_s$ = t.

### Stochastic Taylor Expansion

**Ito‚Äôs Lemma**:- 

Let $f(t,X_t)$ be a stochastic process which satisfies the stochastic differential equation in (1), if $B_t$ is a standard Brownian motion and f is twice differentiable, then $f(t,X_t)$ is also a stochastic process with its differential given by:

$$df(t,X_t)= [\mu(t,X_t)\frac{\partial}{\partial X}f(X(t)))+\frac{\sigma^2(t,X_t)}{2} \frac{\partial^2}{\partial X_t ^2}f(X(t))] dt + \mu(t,X_t)\frac{\partial}{\partial X}f(X(t))dB_t  $$

Ito-Taylor expansion:-
Given an SDE, which is defined as,

$$dX(t)= a[X(t)]dt+b[X(t)]dB(t) \ \ \ \ \  (3) $$ 

By Itos Lemma, we have,
$$df(t,X_t)= [a[X(t)]\frac{\partial}{\partial X}f(X(t)))+\frac{b^2[X(t)]}{2} \frac{\partial^2}{\partial X_t ^2}f(X(t))] dt + a(X(t))\frac{\partial}{\partial X}f(X(t))dB_t  \ \ \ \ \ (4)$$ 

Defining the following operators:-

$$\mathcal{L}^0 = ùëé[X(t)]\frac{\partial}{\partial ùëã}+\frac{b^2}{2}[X(t)]\frac{\partial}{\partial ùëã^2}$$

$$\mathcal{L}^1=ùëè[X(t)]\frac{\partial}{\partial ùëã}$$

Using equation 4, we get,

$$df(X_t)= \mathcal{L}^0 f(X(t))+\mathcal{L}^1 f(X(t))dB_t \ \ \ \ \ (5) $$

By integrating (5) we get,

$$\int_{t_0}^{t} df[(X(t)]= \int_{t_0}^{t}\mathcal{L}^0 f[X(s)] ds+ \int_{t_0}^{t}\mathcal{L}^1 f[X(s)] dB(s)$$


$$f[(X_t)]=f[(X_{t_0})]+ \int_{t_0}^{t}\mathcal{L}^0 f[X(s)] ds+ \int_{t_0}^{t}\mathcal{L}^1 f[X(s)] dB(s) \ \ \ \ \ (6)$$



Now, we are going to make different choices for $f(x)$:

(1) Choose $f(x)$; then (6) becomes:-

$$X(t)=X(t_0)+\int_{t_0}^{t}a[X(s)] ds+ \int_{t_0}^{t}b[X(s)] dB(s)\ \ \ \ \ (7)$$

(2) Choose $f(x)=a(x)$; then (6) becomes:-

$$ùëé[ùëã(ùë°)]=ùëé[ùëã(ùë°0)]+ \int_{t_0}^{t}\mathcal{L}^0 a[X(s)] ds+\int_{t_0}^{t}\mathcal{L}^1 a[X(s)] dB(s) \ \ \ \ \ (8)$$

(3) Choose $f(x)=b(x)$; then (6) becomes:-

$$b[X(t)]=b[X(t_0)]+ \int_{t_0}^{t}\mathcal{L}^0 b[X(s)] ds+\int_{t_0}^{t}\mathcal{L}^1 b[X(s)] dB(s) \ \ \ \ \ (9)$$

Substituting (8) and (9) into (7), we get:-
$$X(t)=X(t_0)+ \int_{t_0}^{t}[ùëé[ùëã(ùë°0)]+ \int_{t_0}^{s_1}\mathcal{L}^0 a[X(s_2)] ds_2+\int_{t_0}^{s_1}\mathcal{L}^1 a[X(s_2)] dB(s_2)] ds_1 +\int_{t_0}^{t}[b[ùëã(ùë°0)]+ \int_{t_0}^{s_1}\mathcal{L}^0 b[X(s_2)] ds_2+\int_{t_0}^{s_1}\mathcal{L}^1 b[X(s_2)] dB(s_2)] ds_1 \ \ \ \ \  \ \ \ \ \ \ \ \ \ \ \ \ \ \  (10)$$


Applying our defined operator, we have:-

$$ \mathcal{L}^0 a[X(s_1)] = a[X(t)]\frac{\partial}{\partial X(s_1} a[X(s_1)] + \frac{b^2[(X(t)]}{2}\frac{\partial}{\partial X^2(s_1)} a[X(s_1)]  $$

$$ \mathcal{L}^0 a[X(s_1)]  = a[X(t)]a'[X(s_1)] + \frac{b^2[X(t)]}{2} a''[X(s_1)] \ \ \ \ \ \ (11) $$

$$ \mathcal{L}^0 b[X(s_1)]ds_1  = a[X(t)]b'[X(s_1)] + \frac{b^2 [(X(t)]}{2} b''[X(s_1)] \ \ \ \ \ \ (12)$$

$$ \mathcal{L}^1 a[X(s_1)] = b[X(t)]a'[X(s_1)] \ \ \ \ \ \ \ \ \ \ (13)$$


$$ \mathcal{L}^1 b[X(s_1)] = b[X(t)]b'[X(s_1)] \ \ \ \ \ \ (14)$$

Thus, 

$$X(t) = X(t_0) + \int_{t_0}^{t} a[X(t_0)] ds_1 + \int_{t_0}^{t}\int_{t_0}^{s_1} \mathcal{L}^0 a[X(s_2)] ds_2] ds_1 \ \ \ \ \ \ (15)$$

$$+\int_{t_0}^{t}\int_{t_0}^{s_1} \mathcal{L}^1 a[X(s_2)] dB[(s_2)] ds_1 + \int_{t_0}^{t} b[X(t_0)] dB(s_1) + \int_{t_0}^{t}\int_{t_0}^{s_1} \mathcal{L}^0 a[X(s_2)] d(s_1) d(B(s_1))] + \int_{t_0}^{t}\int_{t_0}^{s_1} \mathcal{L}^1 b[X(s_2)] dB(s_2) d(B(s_1))]$$

 $$= \boxed{X(t_0) + \int_{t_0}^{t} a[X(t_0)] ds_1 + \int_{t_0}^{t} b[X(t_0)] dB(s_1) + R \ \ \ \ \ \  (16)}$$


where,

$$R= \int_{t_0}^{t}\int_{t_0}^{s_1} \mathcal{L}^0 a[X(s_2)] ds_2 ds_1 +\int_{t_0}^{t}\int_{t_0}^{s_1} \mathcal{L}^1 a[X(s_2)] dB[(s_2)] ds_1 + \int_{t_0}^{t}\int_{t_0}^{s_1} \mathcal{L}^0 a[X(s_2)] d(s_1) d(B(s_1))
+ \int_{t_0}^{t}\int_{t_0}^{s_1} \mathcal{L}^1 b[X(s_2)] dB(s_2) d(B(s_1)) \ \ \ \ \  (17)$$

is the remainder

Simplifying, (16) becomes:-
$$
\begin{align}
X(t)&=X(t_0)+ a[X(t_0)]\int_{t_0}^{t} ds_1 + b[X(t_0)]\int_{t_0}^{t} dW(s_1)+R$ 
\end{align} 
$$

$$
\begin{align}
X(t)=X(t_0)+ \int_{t_0}^{t} a[X(t_0)]ds_1 + \int_{t_0}^{t}b[X(t_0)] dB(s_1)+ b[(X(t)]b'[X(t_0)]\int_{t_0}^{t}\int_{t_0}^{s_1} dB(s_2)dB(s_1)\ \ \ \ \ \space (18)
\end{align}
$$

$$
\begin{align}
\int_{t_0}^{t}\int_{t_0}^{s_1} dB(s_2)dB(s_1) =  \int_{t_0}^{t} B(s_1)dB(s_1)-B(t_0)dB(s_1)
\end{align}
$$

$$
\begin{align}
\int_{t_0}^{t}\int_{t_0}^{s_1} dB(s_2)dB(s_1) =  \frac{[B(t)-B(t_0)]^2}{2}-\frac{t-t_0}{2}
\end{align}
$$

Thus equation (17) becomes,
$$
\begin{align}
X(t)=X(t_0)+ \int_{t_0}^{t} a[X(t_0)]ds_1 + \int_{t_0}^{t}b[X(t_0)] dB(s_1)+ b[(X(t)]b'[X(t_0)][\frac{[B(t)-B(t_0)]^2}{2}-\frac{t-t_0}{2}] + R \ \ \ \ \ (19)
\end{align}
$$

Thus,
$$
\begin{align}
\boxed{X(t)=X(t_0)+ a[X(t_0)]\int_{t_0}^{t}ds_1 + b[X(t_0)]\int_{t_0}^{t} dB(s_1)+ b[(X(t)]b'[X(t_0)][\frac{[B(t)-B(t_0)]^2}{2}-\frac{t-t_0}{2}] + R} \ \ \ \ \ (20)
\end{align}
$$


By discretizing the time $0=t_0<t_1<......<t_N=t$, we get the numerical approximation for the SDE we introduced in equation (3)

$$
\begin{align}
\boxed{X(t_{i+1})=X(t_i)+ a[X(t_i)]\Delta t + b[X(t_i)]\Delta B_i+ b[(X(t_i)]b'[X(t_i)][\frac{[\Delta B_i]^2 - \Delta t}{2}] + R} \ \ \ \ \ (21)
\end{align}
$$

where $\Delta t=t_{i+1}-t_{i}, \Delta B_i = B_{t_{i+1}}-B_{t_{i}}$ for $i=0,1,2, \cdots N-1$ and $X(t_0)=X_0$

      

#### Euler Maruyama scheme

By only including up to the first order terms from equation (21) we get the Euler Maruyama scheme,

$$
\begin{align}
\boxed{X(t_{i+1})=X(t_i)+ a[X(t_i)]\Delta t + b[X(t_i)]\Delta B_i} \ \ \ (22)
\end{align}
$$


#### Milstein scheme

By only including up to the second order terms from equation (21) we get the Milstein scheme,

$$
\begin{align}
\boxed{X(t_{i+1})=X(t_i)+ a[X(t_i)]\Delta t + b[X(t_i)]\Delta B_i+ b[(X(t_i)]b'[X(t_i)][\frac{[\Delta B_i]^2 - \Delta t}{2}]} \ \ \ \ \ (23)
\end{align}
$$


### **Numerical Solution of Black-Scholes SDE**

The Black-Scholes SDE is defined as:-

$$dX(t)= rX(t)dt+\sigma X(t)dBt \ \ \ \ \ (24) $$


where,

$X(t)$ = stock price at time $t$

$r$ = risk-free interest rate

$\sigma$ = volatility or standard deviation of the process

$B_t$ = Standard Brownian motion 

Using equations 20 to 24, we get:-

$$
\begin{align}
\boxed{X(t_{i+1})=X(t_i)+ r[X(t_i)]\Delta t + \sigma[X(t_i)]\Delta B_i+ \sigma^2[X(t_i)][\frac{[\Delta B_i]^2 - \Delta t}{2}] + R} \ \ \ \ \ (25)
\end{align}
$$


```python
import numpy as np
import matplotlib.pyplot as plt
import math
from statistics import mean
```

#### Euler Maruyama

With the aim of computing the expected value in mind, we use the stock price at the expiration date for each of the 10000 iterations to compute the $max(X(T)‚àíK,0)$ in each instance, and use the associated mean in our calculation. Note that the stochastic nature of the problem results in heterogeneous solutions. 

```python
def euler_maruyama(interval,w0,n):
    t= [0]*(n+1)
    w= [0]*(n+1)
    db=[0]*(n+1)
    
    t[0]=interval[0]
    w[0]=w0
    
    h = (interval[1]-interval[0])/n
    sqdelt = h**0.5
    r = 0.05; sigma = 0.25
    
    for i in range(n):
        db[i] = sqdelt*np.random.normal(0,1,1)
        t[i+1] = t[i]+h;
        w[i+1] = w[i]+r*w[i]*h+sigma*w[i]*db[i]
    
    return t,w
```

```python
xvalue = np.zeros(10000)
meanxmatrix = np.zeros(10000)    #initialize matrix for mean
difference = np.zeros(10000)     #initialize matrix for difference
expectedcall = np.zeros(10000)   #initialize matrix for expected value

for j in range(10000):           
    t,w=euler_maruyama([0,.5],12,100)
    xvalue[j] = w[100] #collect the w value at t=.5 for each iteration
    meanxmatrix[j] = 1
    if xvalue[j]>15:
        difference[j] = xvalue[j]-15
    else:
        difference[j] = 0
```

```python
mean_xvalue = np.mean(difference);  
expectedcallvalue = np.exp(-.05*.5)*mean_xvalue  #compute the expected value of the call
meanxmatrix = meanxmatrix*mean_xvalue; #compute a matrix with the mean x value in every entry


fig = plt.figure(figsize=(16,6))
axes = fig.add_subplot(1,2,1)
axes.plot(t,w)
axes.set_xlabel("Time")
axes.set_ylabel("Stock Price")
plt.title("Stock Price vs. Time")

axes = fig.add_subplot(1,2,2)
axes.plot(difference)
axes.plot(meanxmatrix,'ro')
axes.set_xlabel("Iteration Number")
axes.set_ylabel("max(stock price - strike price, 0)")
plt.title("Plot of Call Values vs Iteration Number")
plt.show()
```

To gain an accurate picture of the error associated with the Euler-Maruyama Method, we complete 100 iterations of the approximation for use in our comparison with the closed-form solution.

```python
expectedcallvalue=[0]*100
expectedcallmatrix = np.zeros(100)
actualcallmatrix = np.zeros(100)
imatrix = np.zeros(100)
mm=[0]*100

for i in range(100):           #obtain estimated call values
    xvalue = np.zeros(1000)    #initialize matrix for w values obtained for each iteration at t=.5
    difference = np.zeros(1000)   #initialize matrix for difference
    expectedcall = np.zeros(1000) #initialize matrix for expected value

    for j in range(1000): #iterate through at least 1,000 repititions
        t,w=euler_maruyama([0,0.5],12,100);
        xvalue[j] = w[100] #collect the w value at t=.5 for each iteration
        meanxmatrix[j] = 1
        if xvalue[j]>15:
            difference[j] = xvalue[j]-15
        else:
            difference[j] = 0
    
    mean_xvalue = mean(difference)
    mm[i] = mean_xvalue;
    expectedcallvalue[i] = math.exp(-.05*.5)*mean_xvalue  #compute the expected value of the call
    expectedcallmatrix[i] = expectedcallvalue[i]
    actualcallmatrix[i] = 1
    imatrix[i] = i

    
#Actual Values
actualcallvalue = 12*((1+math.erf(((math.log(12/15)+(.05+.5*(.25**2))*(.5))/(.25*math.sqrt(.5)))/math.sqrt(2)))/2)-(15*math.exp(-.05*.5)*(((1+math.erf(((math.log(12/15)+(.05-.5*(.25**2))*(.5))/(.25*math.sqrt(.5)))/math.sqrt(2)))/2)))
actualcallmatrix = actualcallvalue*actualcallmatrix

mmmatrix = np.mean(mm)*np.ones(100)

absoluteerror = abs(expectedcallmatrix-actualcallmatrix)
absoluteerroreuler = absoluteerror
```

```python
fig = plt.figure(figsize=(16,6))
axes = fig.add_subplot(1,2,1)
axes.plot(imatrix,expectedcallmatrix)
axes.plot(imatrix,actualcallmatrix,'r',label='Actual Solution')
axes.plot(imatrix,mmmatrix,'g',label='Expected Call Value')
axes.legend(loc='best')
plt.title('Expected Call Value vs Iteration Number')
plt.xlabel('Iteration Number')
plt.ylabel('Call Value')

axes = fig.add_subplot(1,2,2)
axes.plot(absoluteerroreuler,'b')
axes.plot(imatrix, [absoluteerroreuler.mean()]*len(imatrix),'r',label='Mean Error')
axes.legend(loc='best')
plt.title('Absolute Error vs. Iteration Number')
plt.xlabel('Iteration Number')
plt.ylabel('Absolute Error')
plt.show()
```

#### Milstein Scheme

```python
def milstein(interval,w0,n):
    t= [0]*(n+1)
    w= [0]*(n+1)
    db=[0]*(n+1)
    t[0]=interval[0]
    w[0]=w0
    
    h = (interval[1]-interval[0])/n
    sqdelt = h**0.5
    r = 0.05; sigma = 0.25
    
    for i in range(n):
        db[i] = sqdelt*np.random.normal(0,1,1)
        t[i+1] = t[i]+h
        w[i+1] = w[i]+r*w[i]*h+sigma*w[i]*db[i]+(.5*(sigma**2)*w[i]*((db[i]**2)-h))
    
    return t,w
```

```python
expectedcallvalue2=[0]*100
expectedcallmatrix2 = np.zeros(100)
actualcallmatrix2 = np.zeros(100)
imatrix = np.zeros(100)

for i in range(100):  #obtain different estimated call values
    xvalue2 = np.zeros(1000) #initialize matrix for w values obtained for each iteration at t=.5
    difference2 = np.zeros(1000)  #initialize matrix for difference
    expectedcall2 = np.zeros(1000) #initialize matrix for expected value

    for j in range(1000): #iterate through at least 1,000 repititions
        [t,w]=milstein([0,0.5],12,100)
        xvalue2[j] = w[100] #collect the w value at t=.5 for each iteration
        if xvalue2[j]>15:
            difference2[j] = xvalue2[j]-15
        else:
            difference2[j] = 0
        
    meanxvalue2 = mean(difference2) 
    expectedcallvalue2[i] = np.exp(-.05*.5)*meanxvalue2  #compute the expected value of the call
    expectedcallmatrix2[i] = expectedcallvalue2[i]
    actualcallmatrix2[i] = 1
    imatrix[i] = i
    i=i+1;

actualcallvalue2 = 12*((1+math.erf(((math.log(12/15)+(.05+.5*(.25**2))*(.5))/(.25*math.sqrt(.5)))/math.sqrt(2)))/2)-(15*math.exp(-.05*.5)*(((1+math.erf(((math.log(12/15)+(.05-.5*(.25**2))*(.5))/(.25*math.sqrt(.5)))/math.sqrt(2)))/2)))
actualcallmatrix2 = actualcallvalue2*actualcallmatrix2

mmmatrix = np.mean(mm)*np.ones(100)
absoluteerror2 = abs(expectedcallmatrix2-actualcallmatrix2)
mm3matrix = mean(expectedcallmatrix2)*np.ones(100)
mm4matrix = mean(absoluteerror2)*np.ones(100)
```

```python
fig = plt.figure(figsize=(16,6))
axes = fig.add_subplot(1,2,1)
axes.plot(imatrix,expectedcallmatrix2)
axes.plot(imatrix,actualcallmatrix2,'r',label='Actual Call Value')
axes.plot(imatrix,mm3matrix,'g',label='Expected Call Value')
plt.title('Expected Call Value vs. Iteration Number with the Actual Call Value Superimposed')
axes.set_xlabel('Iteration Number')
axes.set_ylabel('Call Value')
axes.legend(loc='best')

axes = fig.add_subplot(1,2,2)
axes.plot(imatrix,absoluteerror2)
axes.plot(imatrix,mm4matrix,'r',label='Mean Error')
axes.legend(loc='best')
plt.title('Absolute Error vs. Iteration Number')
plt.xlabel('Iteration Number')
plt.ylabel('Absolute Error')

plt.show()
```

#### Comparison Study

```python
errordiff = absoluteerror-absoluteerror2;  
meanerrormatrix = mean(errordiff)*np.ones(100);  

fig = plt.figure(figsize=(16,6))
axes = fig.add_subplot(1,2,1)

axes.plot(absoluteerror,label='Euler-Maruyama')
axes.plot(absoluteerror2,'r',label='Milstein')
axes.legend(loc='best')
plt.title('Plot of Absolute Errors')
plt.ylabel('Error')
plt.xlabel('Iteration Number')

axes = fig.add_subplot(1,2,2)
axes.plot(imatrix,errordiff,imatrix,meanerrormatrix,'r')
plt.title('Relative Error Between Euler-Maruyama and Milstein Methods')
plt.xlabel('Iteration Number')
plt.ylabel('Error Difference')
plt.show()
```

In comparing the accuracy of the alternative methods, we find that Milstein is more accurate in the few trials we performed, however without a significant number of additional trials it would be myopic to make conclusive probabilistic assertions; in fact, since the Euler-Maruyama and Milstein Methods are SDE solvers, their outputs are non-deterministic, and thus in any finite comparison of their accuracy it is possible for either method to outperform the other.
